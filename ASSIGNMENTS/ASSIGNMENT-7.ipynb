```python
 #1 Datastring to timeseries
import pandas as pd
dates=['2020-01-01','2020-01-02','2020-01-03','2020-01-04']
ts=pd.to_datetime(dates)
print(ts)
```

```python
d1=pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Aman','Riya','Karan','Priya']
})
d2=pd.DataFrame({
    'id':[2,3,4,5],
    'marks':[85,90,78,80]
})
print(d1)
print(d2)
```

```python
#INNER MERGE
res=d1.merge(d2,on='id',how='inner')
print("INNER MERGE")
print(res)
```

```python
print("Left Merge")
r1=d1.merge(d2,on='id',how='left')
print(r1)
```

```python
print("right merge")
r2=d1.merge(d2,on='id',how='right')
print(r2)
```

```python
print("index based join ")
r3=d1.join(d2,lsuffix='_left',rsuffix='_right')
print(r3)
```

```python
#merge on multiple keys
d3=pd.DataFrame({
    'id':[1,2,3],
    'dept':['IT','HR','IT'],
    'salary':[50000,45000,60000]
})
d4=pd.DataFrame({
    'id':[1,2,3],
    'dept':['IT','HR','Finance'],
    'bonus':[5000,4000,7000]
})
print("Multiple Merge")
r5=d3.merge(d4,on=['id','dept'],how='inner')
print(r5)
```

```python
# 3
d6=pd.DataFrame({
    'id':[1,2],
'name':["Aman","Riya"]
})
d7=pd.DataFrame({
    'id':[3,4],
    'name':["Karan","Priya"]
})
d8=pd.DataFrame({"id":[1,2,3,4],
                 'marks':[85,82,30,94]})
print("Concatenamtion\n")
r1=pd.concat([d6,d7])
print(r1)
r2=r1.merge(d8,on='id')
print("Merged with third DataFrame")
print(r2)
```

```python
print("join() : joins on indexes by default \n merge(): joins on columns by default")
print("MERGE")
r3=d6.merge(d8,on='id')
print(r3)
print("JOIN")
r4=d6.join(d8,lsuffix="_left",rsuffix="_right")
print(r4)
```

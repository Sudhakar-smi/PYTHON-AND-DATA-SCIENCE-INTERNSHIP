```python
#1) Create a CSV file for address book, CSV file should have column for name, address, mobile, email. Insert 2-3 dummy data entered by user.

import csv
f=open('address_book.csv', 'w', newline='')
writer = csv.writer(f)

writer.writerow(['Name','Address','Mobile','Email'])

n=int(input("Enter number of entries: "))
for i in range(n):
    name=input("Enter name:")
    address=input("Enter address:")
    mobile=input("Enter mobile")
    email=input("Enter email:")

    writer.writerow([name,address,mobile,email]) 


f.close()


f=open('address_book.csv', 'r')
reader=csv.reader(f)
for row in reader:
    print(row)
```

```python
# 2)  Practice DATABASE  
# 1) Create Database
# 2) Create 2-3 tables
# 3) Insert some records
# 4) Perform diffrent select operations
# 5) Update some data
# 6) Delete some data

import sqlite3
conn=sqlite3.connect('Class3.db')

conn.execute("""
             CREATE TABLE IF NOT EXISTS Student(
             RollNo INTEGER PRIMARY KEY,
             Name VARCHAR(20),
             Age INTEGER)
             """)
conn.execute("""CREATE  TABLE IF NOT EXISTS Course(
             CourseID integer PRIMARY KEY,
             CourseName VARCHAR(20))
             """)

conn.execute("INSERT INTO Student VALUES(1,'Shekahr',20)")
conn.execute("INSERT INTO Student VALUES(2,'Sonam',20)")
conn.execute("INSERT INTO Student VALUES(3,'Sumit',20)")
conn.commit()

conn.execute("INSERT INTO Course VALUES(101,'Python')")
conn.execute("INSERT INTO Course VALUES(102,'Java')")
conn.execute("INSERT INTO Course VALUES(103,'C++')")
conn.commit()

res=conn.execute("select* from Student")
for row in res:
    print(row)

res=conn.execute("select* from Course")
for row in res:
    print(row)

res=conn.execute("select * from Student where RollNo = 1")
for row in res:
    print(row)

res=conn.execute("select * from Course where CourseName = 'Python'")   
for row in res:
    print(row)


conn.execute("update student set Age = 21 where name = 'Sonam'") 
conn.execute("update course set CourseName = 'Python Programming' where CourseID = 101")
conn.commit()


res=conn.execute("select * from Student")
for row in res:
    print(row)

res=conn.execute("select * from Course")
for row in res:
    print(row)    


conn.execute("delete from student where name ='sumit'")
conn.commit()

res=conn.execute("select * from Student")
for row in res:
    print(row)


conn.close()    
```

### b) Selecting rows in pandas DataFrame based on conditions

```python
import pandas as pd

data = {
    "name": ["SHEKHAR", "SUDHAKAR", "MOHIT"],
    "age": [21, 18, 25]
}

df = pd.DataFrame(data)

result = df[df["age"] > 20]

print(result)
```

### c) Select any row from a DataFrame using iloc[]

```python
import pandas as pd

data = {
    "name": ["SHEKHAR", "SUDHAKAR", "MOHIT"],
    "age": [21, 18, 25]
}

df = pd.DataFrame(data)



print(df.iloc[1])
```

### d) Limited rows selection with given column

```python
import pandas as pd

data = {
    "name": ["SHEKHAR", "Sudhakar", "Mohit"],
    "age": [21, 22, 23],
    "City": ["SIKAR", "NIWAI", "JAIPUR"]
}

df = pd.DataFrame(data)

print(df.loc[0:1, ["name", "City"]])
```

### e) Drop rows from the DataFrame based on condition

```python
import pandas as pd

data = {
    "name": ["SHEKHAR", "Sudhakar", "Mohit"],
    "age": [21, 22, 23],
    "City": ["SIKAR", "NIWAI", "JAIPUR"]
}

df = pd.DataFrame(data)

df=df[df["name"]=="SHEKHAR"]
print(df)
```

```python
# 1) Replace Nan with 0 and Interchange 3 rows and 3 columns of 2D array [[6, -8, 73, -110], [np.nan, -8, 0, 94]]

import numpy as np

a = np.array([[6, -8, 73, -110],
            [np.nan, -8, 0, 94]])

a = np.nan_to_num(a, nan=0)
print("NaN replace with 0:\n", a)

interchange_a = a.T
print("Rows and columns interchanged:\n", interchange_a)
```

```python
# 2) Move axes of 3D array to new positions 

import numpy as np

a3d = np.arange(24).reshape(2, 3, 4)

print("Original shape:", a3d.shape)

new_a = np.moveaxis(a3d, 0, 2)

print("After moving axis:", new_a.shape)
print(new_a)
```

```python
 # 3)Replace NaN values with average of columns

import numpy as np

a = np.array([[1, 2, np.nan],
                [4, np.nan, 6],
                [7, 8, 9]])

col_mean = np.nanmean(a, axis=0)

indx = np.where(np.isnan(a))
a[indx] = np.take(col_mean, indx[1])

print(a)
```

```python
# 4) Replace negative value with zero in numpy array using replace
import numpy as np

a = np.array([6, -8, 73, -110, 0, 94])

a = np.where(a < 0, 0, a)

print(a)
```

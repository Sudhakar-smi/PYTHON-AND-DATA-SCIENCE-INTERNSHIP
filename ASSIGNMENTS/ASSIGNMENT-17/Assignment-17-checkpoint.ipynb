```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("SalesData") \
    .getOrCreate()

print("Spark Started")
```

```python
#Read CSV
df = spark.read.csv(
    "sales.csv",
    header=True,
    inferSchema=True
)

df.show()
```

```python
#Sort by Sales Descending
sorted_data = df.orderBy(
    "sales",
    ascending=False
)

sorted_data.show()
```

```python
#Top 3 Products
top3 = sorted_data.limit(3)

top3.show()
```

```python
#Sales Greater Than 80000
result = df.filter(
    df.sales > 80000
)

result.show()
```

```python
#Save Output CSV
result.write.mode("overwrite").csv(
    "output"
)

print("File Saved")
```

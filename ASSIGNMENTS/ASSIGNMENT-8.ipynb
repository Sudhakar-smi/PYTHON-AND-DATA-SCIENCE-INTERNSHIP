```python
import pandas as pd
df = pd.DataFrame({
    'Date':pd.to_datetime(['2026-05-20', '2026-05-21', '2026-05-22']*4),
    'Sales': [100, 200, 150,250,300,350,400,450,500,550,600,650]
})
print(df)
df['Year']=df['Date'].dt.year
df['Month']=df['Date'].dt.month_name()
df['day']=df['Date'].dt.day_name()
print(df)
```

```python
import pandas as pd
d1=pd.read_csv("tip.csv")
day_total=d1.groupby('day')['total_bill'].sum()
print("Day wise total bill:")
print(day_total)
```

```python
d1=pd.read_csv("tip.csv")
time_total=d1.groupby('time')['total_bill'].sum()
print("time wise total bill:")
print(time_total)
```

```python
d1=pd.read_csv("tip.csv")
tip_total=d1.groupby('tip')['total_bill'].sum()
print("TIP wise total bill:")
print(tip_total)
```

```python
d1=pd.read_csv("tip.csv")
day_max=d1.groupby('day')['total_bill'].max()
print("Day wise Max bill:")
print(day_max)
```

```python
d1=pd.read_csv("tip.csv")
day_count=d1.groupby('day')['total_bill'].count()
print("Day wise count bill:")
print(day_count)
```

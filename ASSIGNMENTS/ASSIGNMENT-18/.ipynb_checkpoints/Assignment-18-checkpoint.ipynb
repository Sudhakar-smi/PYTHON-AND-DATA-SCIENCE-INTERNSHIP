```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PartitionDemo") \
    .getOrCreate()

print("Spark Started")
```

```python
#5 million rows wala DataFrame
df = spark.range(5000000)

df.show(5)
```

```python
print("Initial Partitions :")

print(df.rdd.getNumPartitions())
```

```python
#Repartition to 12
df2 = df.repartition(12)

print("After Repartition :")

print(df2.rdd.getNumPartitions())
```

```python
#Coalesce to 3
df3 = df2.coalesce(3)

print("After Coalesce :")

print(df3.rdd.getNumPartitions())
```

```python
print("Initial Partitions :", df.rdd.getNumPartitions())
print("After Repartition :", df2.rdd.getNumPartitions())
print("After Coalesce :", df3.rdd.getNumPartitions())
```

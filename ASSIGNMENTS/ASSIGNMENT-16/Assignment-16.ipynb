```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("EmployeeData") \
    .getOrCreate()

sc = spark.sparkContext

print("Spark Started")
```

```python
#CSV Read

rdd = sc.textFile("employee.csv")
header = rdd.first()
employees = rdd.filter(
    lambda x: x != header
).map(
    lambda x: x.split(",")
)

employees.collect()
```

```python
#Salary Descending
sorted_emp = employees.sortBy(
    lambda x: int(x[3]),
    ascending=False
)

sorted_emp.collect()
```

```python
#Display Sorted Employees

print("Employees Sorted By Salary")

for i in sorted_emp.collect():
    print(i[1], i[3])
```

```python
#Department Wise Total Salary
dept = employees.map(
    lambda x: (x[2], int(x[3]))
)

total = dept.reduceByKey(
    lambda a, b: a + b
)

total.collect()
```

```python
#Display Department Totals

print("Department Wise Total Salary")

for i in total.collect():
    print(i[0], i[1])
```

```python
#Top 3 Employees

top3 = sorted_emp.take(3)

top3
```

```python
#Display Top 3
print("Top 3 Highest Paid Employees")

for i in top3:
    print(i[1], i[3])
```

```python
#Save Output File
file = open("top3.txt", "w")

for i in top3:
    file.write(i[1] + " " + i[3] + "\n")

file.close()

print("File Saved")
```

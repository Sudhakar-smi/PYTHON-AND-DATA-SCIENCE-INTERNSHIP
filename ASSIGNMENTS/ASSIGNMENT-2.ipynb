```python
# 2) Write a function for basic math operations like add multiply substract divide and use this in your program, take 2 number input from user.

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b != 0:
        return a / b
    else:
        return "Division by zero is not possible"



n1 = float(input("Enter first number: "))
n2 = float(input("Enter second number: "))

print("Addition =", add(n1, n2))
print("Subtraction =", subtract(n1, n2))
print("Multiplication =", multiply(n1, n2))
print("Division =", divide(n1, n2))
```

```python

 #3) Write a program to check Palindrome Number For example Number 12321 is a Palindrome Number, because 12321 is equal to its reverse Number 12321. Steps for checking Palindrome number

 #1. Find reverse of the given number.
 #2. Compare that number with the reverse number.
 #3. If number and its reverse is equal then it is a Palindrome Number otherwise not.

n=input("Enter a number: ")
reverse_n=n[::-1]
if n==reverse_n:
    print(n, "is a palindrome.")
else:
    print(n, "is not a palindrome.")
```

```python
# 1) Practise
# 	Dictionary

print("DICTIONARY")
student={
    "name":"John",  
    "age": 20,
    "major": "Computer Science"}
print(student)
print("Accessing values:")
print("Name:", student["name"])
print("Age:", student["age"])
print("ADDING NEW KEY-VALUE PAIR")
student["year"]="2nd_year"
print(student)
print("UPDATING EXISTING KEY-VALUE PAIR")
student["age"]=21
print(student)
print("DELETING A KEY-VALUE PAIR")
del student["major"]
print(student)
print("student.keys():", student.keys())
print("student.values():", student.values())
print("student.items():", student.items())



```

```python
#1) Practise
#	Tuple


numbers=(1,2,3,3,3,4,5)
print(numbers)
print("Accessing values:")
print("First number:", numbers[0])
print("Slicing:", numbers[1:4])
print("length:", len(numbers))
print("Count of 3:", numbers.count(3))
print("Index of first occurrence of 3:", numbers.index(3))
print("sum:", sum(numbers))

```

```python
# 1) Practise
#	Set

fruits={"apple", "banana", "orange", "apple", "grape"}
print(fruits)
print("Adding a fruit:")
fruits.add("kiwi")
print(fruits)
print("Removing a fruit:")
fruits.remove("banana")
fruits2={"grape", "melon", "kiwi"}
print("Union:", fruits.union(fruits2))
print("Intersection:", fruits.intersection(fruits2))
print("Difference:", fruits.difference(fruits2))
print("grapes in fruits but not in fruits2:", fruits.difference(fruits2))
```
